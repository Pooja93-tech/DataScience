```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('NetFlix Data.csv',encoding='latin-1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
X= df.drop(['ID','Title','ReleaseDate','VideoReleaseDate','IMDB'],axis=1)
```

```python
X.head()
```

```python
X.describe()
```

```python
X.shape
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
wardlink = linkage(X, method = 'ward')
dend = dendrogram(wardlink,
                 truncate_mode='lastp',
                 p = 10,
                 )
```

```python
from scipy.cluster.hierarchy import fcluster
clusters = fcluster(wardlink, 6, criterion='maxclust')
clusters
```

```python
df['clusters'] = clusters
df.head()
```

```python
df.clusters.value_counts().sort_index()
```

```python
Genre=df.iloc[:,5:24]
```

```python
df.groupby('clusters')[Genre.columns].sum()
```
